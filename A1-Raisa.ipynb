```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.DataFrame(
    np.random.rand(10,4),
    columns=['a','b','c','d']
)

df.plot.area()
plt.title("Random Data Visualization")
plt.xlabel("Index")
plt.ylabel("Values")
plt.show()
```
